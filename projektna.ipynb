V tej nalogi sem analiziral statistiko iger v igrici Age Of Empires 2, bolj podrobno igre 1 proti 1.
Za podatke sem si izbral spletno stran https://aoestats.io/. Prvo kot prvo bomo prenesli tabele, ki smo jih izluščili iz spletne strani.  

```python
import pandas as pd
glavni_podatki = pd.read_csv('civilizacije_tabela.csv')
sicilijanci_zeleni = pd.read_csv('sicilijanci_zeleni_nasprotniki.csv')
sicilijanci_nezeleni = pd.read_csv('sicilijanci_nezeleni_nasprotniki.csv')
sicilijanci_otvoritve = pd.read_csv('sicilijanci_otvoritve.csv')
sicilijanci_mape = pd.read_csv('sicilijanci_mape.csv')
megrandom = pd.read_csv("megarandom_tabela.csv")
```

## Glavni Podatki

Ogljemo si prvo tabelo, ki zadrzuje glavne podatke. Prvotno so razvrščeni po zmagovalni stopnji, nato pa igralni.

```python
glavni_podatki
```

## Izbrana civilizacija: Sicilijanci

Sedaj si bomo bolj podrobno ogledali izbrano civilizacijo: Sicilijanci. Kot civilizacija imajo močno pehoto, ter konjenico. Ležijo jim vsi tipi map, saj imajo tudi solidno pomorstvo. Najbolj znani so po svoji edinstveni enoti "Sergeant", ki lahko gradi manjše utrdbe. Za začetek si poglejmo njihove najbolj in najmanj želene nasprotnike.

```python
graf2 = pandica.civ_zeleni(sicilijanci_zeleni)
graf3 = pandica.civ_nezeleni(sicilijanci_nezeleni)
graf2 and graf3
```

Opazimo močno prevlado nad hindunstansko civilizacijo ter strah in trepet pred Rimljani. Oglejmo si sedaj tipične sicilijanske začetne strategije:

```python
sicilijanci_otvoritve
```

Grafični prikaz igralne stopnje posameznih otvoritev:

```python
graf4 = pandica.civ_otvoritve_igr(sicilijanci_otvoritve)
graf4
```

Opazimo, da je najbolj popularno začeti z konjenico "scouts". Kar se tiče vodnih začetkov ("fires" in "galleys"), so zelo redki saj je popularnost vodnih map veliko manjša od zemljskih. Sedaj še uspešnost posameznih otvoritev:

```python
graf5 = pandica.civ_otvoritve_zmg(sicilijanci_otvoritve)
graf5
```

Čeprav je konjenica bolj popularna, imamo večjo možnost za zmago, če izberemo začetno strategijo "fast castle". Sedaj pa še podatki, kjer bomo videli najbolj izbrane mape sicilijanskih igralcev.

```python
sicilijanci_mape
```

```python
graf6 = pandica.civ_mape(sicilijanci_mape)
graf6
```

Mapi "Enclosed" in "Coastal forest" imata premalo iger, da bi ju resno jemali. Zato lahko trdimo, da imamo največjo možnost za zmago, če izberemo mapo "Ghost lake".
