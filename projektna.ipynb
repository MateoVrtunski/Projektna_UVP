# ANALIZA PODATKOV IGRE AGE OF EMPIRES 2

V tej nalogi sem analiziral statistiko iger v igrici Age Of Empires 2, bolj podrobno igre 1 proti 1.
Za podatke sem si izbral spletno stran https://aoestats.io/. Prvo kot prvo, bomo prenesli tabele, ki so bile izluščene iz spletne strani. 
Analizirali bomo glavne podatke in eno od 43ih civilizacij ter mapo po lastni izbiri. 

```python
import pandas as pd
glavni_podatki = pd.read_csv('civilizacije_tabela.csv')
sicilijanci_zeleni = pd.read_csv('sicilijanci_zeleni_nasprotniki.csv')
sicilijanci_nezeleni = pd.read_csv('sicilijanci_nezeleni_nasprotniki.csv')
sicilijanci_otvoritve = pd.read_csv('sicilijanci_otvoritve.csv')
sicilijanci_mape = pd.read_csv('sicilijanci_mape.csv')
megrandom = pd.read_csv("megarandom_tabela.csv")
```

## Glavni Podatki

Ogljemo si prvo tabelo, ki zadržuje glavne podatke. Prvotno so razvrščeni po zmagovalni stopnji, nato pa igralni (v procentih seveda).

```python
glavni_podatki
```

```python
glavni2 = glavni_podatki.sort_values('igralna stopnja',ascending = False)
glavni2
```

Številke sem pustil ob strani, da lahko na hitro primerjamo igralno stopnjo z zmagovalno. Sedaj pa poglejmo graf, ki nam vse to bolj nazorno prikaže.

```python
import pandica
graf1 = pandica.graf_glavni(glavni_podatki)
graf1
```

Opazimo močno prevlado v popularnosti na strani Frankov, ter največjo zmagovalno stopnjo na strani Rimljanov. Tu pa vidimo, da je igra zelo dobro narejena, saj nobena civilizacija ne izstopa preveč v zmagah oz. izgubah razen mogoče Kitajcev.

## Izbrana civilizacija: Sicilijanci

Sedaj si bomo bolj podrobno ogledali izbrano civilizacijo: Sicilijanci. Kot civilizacija imajo močno pehoto, ter konjenico. Ležijo jim vsi tipi map, saj imajo tudi solidno pomorstvo. Najbolj znani so po svoji edinstveni enoti "Sergeant", ki lahko gradi manjše utrdbe. Za začetek si poglejmo njihove najbolj in najmanj želene nasprotnike.

```python
graf2 = pandica.civ_zeleni(sicilijanci_zeleni)
graf3 = pandica.civ_nezeleni(sicilijanci_nezeleni)
graf2 and graf3
```

Opazimo močno prevlado nad hindunstansko civilizacijo ter strah in trepet pred Rimljani. Sedaj pas nas zanima, katere so njihove tipične začetne strategije.

```python
sicilijanci_otvoritve
```

Grafični prikaz igralne stopnje posameznih otvoritev:

```python
graf4 = pandica.civ_otvoritve_igr(sicilijanci_otvoritve)
graf4
```

Opazimo, da je najbolj popularno začeti z konjenico ("scouts"). Kar se tiče vodnih začetkov ("fires" in "galleys"), so zelo redki saj je popularnost vodnih map veliko manjša od zemljskih. Sedaj še uspešnost posameznih otvoritev:

```python
graf5 = pandica.civ_otvoritve_zmg(sicilijanci_otvoritve)
graf5
```

Čeprav je konjenica bolj popularna, imamo večjo možnost za zmago, če izberemo začetno strategijo "fast castle". Na vrsti so še podatki, kjer bomo videli najbolj izbrane mape sicilijanskih igralcev.

```python
sicilijanci_mape
```

```python
graf6 = pandica.civ_mape(sicilijanci_mape)
graf6
```

Mapi "Enclosed" in "Coastal forest" imata premalo iger, da bi ju resno jemali. Zato lahko trdimo, da imamo največjo možnost za zmago, če izberemo mapo "Ghost lake".

## Izbrana mapa: Megarandom

Sedaj pa še podrobneje si oglejmo statistiko izbrane mape: Megarandom. Je najbolj nepredvidljiva mapa, kot že samo ime pove. Ko poženemo mapo Megarandom ne vemo še, ali bomo začeli na otoku, v gorah, puščavi in ali bomo imeli že vojsko oz. celo že zgrajene stavbe. Vemo le eno: igra bo zelo pestra. 

Začetna statistika:

```python
megrandom
```

```python
round(megrandom['igralna stopnja'].mean(),2)
```

Vidimo, da je popularna mapa, saj jo v povprečju ljudje izberejo kar 8.94% časa. Poglejmo si še graf:

```python
graf7 = pandica.mapa_vse(megrandom)
graf7
```

Mapo imajo najraje Gurjarasi, najboljši na njej pa so Huni. Ni pa nobne civilizacije, ki bi znatno izstopala.
